```python
# reading excel file and saving fifth column in list

import pandas as pd
import copy

df = pd.read_excel(r'T8-T6-Experiment.xlsx', sheet_name=None ) #header=None, # usecols="A"
sheet_name= 'B'
# print(df)
# df['ctseriesid']
# df.iloc[:, 0]
df = df[sheet_name]
fifth_column_list = df.iloc[:, 4].tolist()
print(fifth_column_list)

```

```python
# Fining indicies of interest

# Initialize an empty list to store the indices
indices_with_pbv = []

# Find indices of elements containing "PBV"
for index, element in enumerate(fifth_column_list):
    if "PBV" in element:
        indices_with_pbv.append(index)

print(indices_with_pbv)
```

```python
# taking the values at those ndices and saving inotn an new sheet

# Filter elements from "a" based on indices in "i"
filtered_strings = [fifth_column_list[index] for index in indices_with_pbv]

# Create a DataFrame from the filtered strings
df2 = pd.DataFrame({'Filtered_Strings': filtered_strings})

# Replace 'output.xlsx' with the desired Excel file name
output_file = 'T8-T6-Experiment.xlsx'

# Save the DataFrame to a new sheet in the Excel file
with pd.ExcelWriter(output_file, engine='openpyxl', mode='a') as writer:
    df2.to_excel(writer, sheet_name='D', index=False)
    # writer.save()

print("Filtered data saved to the Excel file.")
```

```python
# Find all Patient IDs for PBV Images

# Filter the values from the 2nd column based on the list of indices
selected_values = df.iloc[indices_with_pbv, 3].tolist()

# Print the selected values
print(selected_values)
```
