```python
#8FE Create/update 'edge_length' property for all edges *

def UpdateEdgeLengths(G):
    for edge in G.edges:
        node1, node2 = edge
        x1, y1, z1 = node1
        x2, y2, z2 = node2
        edge_length = np.sqrt((x2 - x1)**2 + (y2 - y1)**2 + (z2 - z1)**2)
        G.edges[edge]['edge_length'] = edge_length

UpdateEdgeLengths(graph)
```

```python
#10F Function to map NewAverageNode to nearest voxel on skeleton, if it doesn't lie on it:

def FindNearestVoxelOnSkeletonIfNotAlreadyOnIt(skeleton, NewAverageNode):

    if skeleton[*NewAverageNode] == 0:
        # print(f"NewAverageNode {NewAverageNode} not on the skeleton.")
        xxna, yyna, zzna = NewAverageNode
        success = False
        NeighbrhdSize = 3
        while success == False:
            # print(f"Searching neighborhood {NeighbrhdSize}")
            lowerbnd = -(NeighbrhdSize//2)
            upperbnd =  (NeighbrhdSize//2) + 1
            neighbors = [(i, j, k) for i in range(lowerbnd, upperbnd) for j in range(lowerbnd, upperbnd) for k in range(lowerbnd, upperbnd)]
            neighbors.remove((0,0,0))
            for neighbor_offset in neighbors:
                nnx, nny, nnz = xxna + neighbor_offset[0], yyna + neighbor_offset[1], zzna + neighbor_offset[2]
                if 0 <= nnx < skeleton.shape[0] and 0 <= nny < skeleton.shape[1] and 0 <= nnz < skeleton.shape[2]:
                    if skeleton[nnx, nny, nnz] != 0:
                        NewAverageNode = (nnx, nny, nnz)
                        # print(f"Adjusted the NewAverageNode to {NewAverageNode}")
                        success = True
                        break            
            NeighbrhdSize += 2

    return NewAverageNode
```

```python
#P1 Function: Find route between start and end, on the skeleton *: Nothing to do with graph/digraph:

import time

def find_route(skeletonn, start, end, EdgeNum):

    visited = set()
    stack = [(start, [start])]  # Each element in the stack is a tuple (current_point, path_so_far)

    # Define possible movements in the 3D space (26 neighbors)
    directions = [(dx, dy, dz) for dx in [-1, 0, 1] for dy in [-1, 0, 1] for dz in [-1, 0, 1] if (dx, dy, dz) != (0, 0, 0)]

    while stack:

        current_point, path_so_far = stack.pop()

        # If we've reached the end, return the path
        if current_point == end:
            num_edges = nx.number_of_edges(graph)
            print(f"Edge #{EdgeNum:3}/{num_edges}; Successfully got from {start} to {end}; TimeElapsed: {(time.time() - StartTime):07.2f}")
            for vxl in path_so_far[5:-5]: # Excluding start and end, set the rest to zero to speed next iteration up
                skeletonn[*vxl] = 0 # Gold! **
            return path_so_far

        # Mark the current point as visited
        visited.add(current_point)

        # Check the 26 neighboring points
        for dx, dy, dz in directions:

            x, y, z = current_point
            neighbor = (x + dx, y + dy, z + dz)

            # Check array bounds and if the neighbor is valid (1-valued and not visited)
            if any(coord < 0 or coord >= dim for coord, dim in zip(neighbor, skeletonn.shape)): continue
            if neighbor in visited or skeletonn[neighbor] == 0: continue

            # Add neighbor to stack along with the updated path
            new_path = path_so_far + [neighbor]
            stack.append((neighbor, new_path))

    # If we get here, there's no valid path from start to end
    num_edges = nx.number_of_edges(graph)
    print(f"Edge #{EdgeNum:3}/{num_edges}; Error! ******* Didn't get from {start} to {end} ******* ")
    return None
```

```python
#P2 Upd-MainFunction: Creates and informs the digraph: Calls traverse function above: $$$ root = () or Enter

# Fucntion: fully traverse the directed tree visiting all edges:
def traverse_full_tree(digraph, current_node):
    global EdgeNum
    for successor in digraph.successors(current_node): # Find and process the outgoing edges from the current node
        visitedvoxels = find_route(skeletonn, current_node, successor, EdgeNum) # Includes the source and target nodes
        digraph[current_node][successor]['LocalPath'] = visitedvoxels
        digraph[current_node][successor]['AssignedEdgeNumber'] = EdgeNum
        EdgeNum += 1
        traverse_full_tree(digraph, successor) # Recursively traverse the tree from the successor node

# Main: Start the traversal from a specified root node:
import time
winsound.Beep(frequency, 20); winsound.Beep(frequency, 150)
user_input = input("If root was OK just press Enter; if not, enter the root: eg (100, 20, 18): [Cell-24>>Cell-16]: ")
if user_input.strip() == "": pass
else: root = eval(user_input)
digraph = nx.dfs_tree(graph, source=root) # Create a directed graph starting from the specified root node
StartTime = time.time()
EdgeNum = 1
skeletonn = copy.deepcopy(skeleton)
traverse_full_tree(digraph, root)
print('\nDone.')
```

```python
#S1 Basic Strahler Analysis: Function+Call: Using DiGraphs: *

import networkx as nx

def StrahlerAnalysis(digraph): # Returns the edges
    
    # Find the terminal edges (edges with no successors)
    terminal_edges = [edge for edge in digraph.edges() if not any(digraph.successors(edge[1]))]
    # Initialize the BscStrahlerOrder for terminal edges as 1
    for edge in digraph.edges():
        if edge in terminal_edges: digraph[edge[0]][edge[1]]['BscStrahlerOrder'] = 1
        else: digraph[edge[0]][edge[1]]['BscStrahlerOrder'] = 0

    # Perform Strahler analysis starting from terminal edges
    while True:

        # Find all edges with the same BscStrahlerOrder
        edges_by_order = {}
        for edge in digraph.edges():
            BscStrahlerOrder = digraph[edge[0]][edge[1]].get('BscStrahlerOrder', 0)
            if BscStrahlerOrder not in edges_by_order:
                edges_by_order[BscStrahlerOrder] = []
            edges_by_order[BscStrahlerOrder].append(edge)

        # If no more BscStrahlerOrder 0, break
        if 0 not in edges_by_order: break

        # Only process edges with BscStrahlerOrder = 0
        for edge in edges_by_order[0]:
            successors = list(digraph.successors(edge[1]))
            SuccessorOrders = [digraph[edge[1]][successor]['BscStrahlerOrder'] for successor in successors]
            if not all(SuccessorOrders): continue # If any successor BscStrahlerOrder is 0, skip to next edge
            if len(successors) == 1: # If only one successor, inherit its BscStrahlerOrder
                digraph[edge[0]][edge[1]]['BscStrahlerOrder'] = digraph[edge[1]][successors[0]]['BscStrahlerOrder']
            else: # So it's a bifurcation
                MaxOrder = max(SuccessorOrders)
                BinaryComparison = [int(ordr == MaxOrder) for ordr in SuccessorOrders]
                # If more than 2 of the max then increment BscStrahlerOrder
                if sum(BinaryComparison)>=2: NextOrder = MaxOrder + 1
                else: NextOrder = MaxOrder
                digraph[edge[0]][edge[1]]['BscStrahlerOrder'] = NextOrder

    return digraph.edges(data=True)

# Main:
EdgesResult = StrahlerAnalysis(digraph)
# for edge in EdgesResult:
#     print(f"Edge: {edge[0]} -> {edge[1]}, Order: {edge[2]['BscStrahlerOrder']}")
```

```python
# S5 Enhance the orders: Diameter-defined Strahler

import networkx as nx
import numpy as np

# Function to calculate new bounds
def calculate_bounds(orders, avg_diam, std_diam):
    bounds = {}
    for order in orders:
        prev_order = order - 1
        next_order = order + 1
        # Calculate left bound
        left_bound  = (avg_diam[order] + avg_diam[prev_order] + std_diam[prev_order] - std_diam[order]) / 2
        # Calculate right bound
        right_bound = (avg_diam[order] + avg_diam[next_order] + std_diam[order] - std_diam[next_order]) / 2
        bounds[order] = (left_bound, right_bound)
    return bounds

# Function to update the orders based on the calculated bounds
def update_orders(digraph, bounds):
    changes = False
    for u, v, data in digraph.edges(data=True):
        order = data['DdsStrahlerOrder']
        diameter = data['AvgDiameterForMid3rdMm']
        
        # Find the new order
        for new_order, (new_left, new_right) in bounds.items():
            if new_left <= diameter <= new_right:
                if new_order != order:
                    changes = True
                data['DdsStrahlerOrder'] = new_order
                break
    return changes


# Starts here
for edge in digraph.edges:
    digraph.edges[edge]['DdsStrahlerOrder'] = digraph.edges[edge]['BscStrahlerOrder']

# Find the range of orders dynamically
all_orders = {data['DdsStrahlerOrder'] for _, _, data in digraph.edges(data=True)}
min_order = min(all_orders)
max_order = max(all_orders)
orders = list(range(min_order, max_order + 1))

alldiameters = [data['AvgDiameterForMid3rdMm'] for u, v, data in digraph.edges(data=True)]
print(f"Diameters range: {min(alldiameters)} - {max(alldiameters)}\n")

RoundNum = 1
while True:
    diameters_by_order = {}
    for order in orders:
        diameters = [data['AvgDiameterForMid3rdMm'] for u, v, data in digraph.edges(data=True) if data['DdsStrahlerOrder'] == order]
        diameters_by_order[order] = diameters
    
    # Calculate average and standard deviation
    avg_diam = {order: np.mean(diam) for order, diam in diameters_by_order.items()} # if diam else 0
    std_diam = {order: np.std (diam) for order, diam in diameters_by_order.items()} # if diam else 0
    
    # Handle edge cases for min_order - 1 and max_order + 1
    avg_diam[min_order - 1] = min([d for diam in diameters_by_order.values() for d in diam])
    avg_diam[max_order + 1] = max([d for diam in diameters_by_order.values() for d in diam])
    std_diam[min_order - 1] = std_diam[max_order + 1] = 0
    
    # Calculate bounds
    bounds = calculate_bounds(orders, avg_diam, std_diam)
    # print("Round", f"{RoundNum: >2}", "Bounds: ", end="")

    # for key, value in bounds.items():
    #     formatted_value = tuple(f"{x:.2f}" for x in value)
    #     print(f"{key}: {formatted_value}")
    formatted_dict_str = '{'
    for key, value in bounds.items():
        formatted_value = ', '.join(f"{x:.2f}" for x in value)
        formatted_dict_str += f"{key}: ({formatted_value}), "
    formatted_dict_str = formatted_dict_str.rstrip(', ') + '}'
    print(formatted_dict_str)

    # Update orders and check for changes
    if not update_orders(digraph, bounds): break
    RoundNum += 1

print(); BscOrderCounts = GetOrderCounts("BscStrahlerOrder")
print(); DdsOrderCounts = GetOrderCounts("DdsStrahlerOrder")
```
